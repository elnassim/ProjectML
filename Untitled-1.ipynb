```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
df = pd.read_csv("dataset_revenu_marocains.csv")

# Calcul des moyennes
mean_global = df['income'].mean()
mean_urbain = df[df['area'] == 'urbain']['income'].mean()
mean_rural = df[df['area'] == 'rural']['income'].mean()

# Préparer les résultats pour le graphique
areas = ['Global', 'Urbain', 'Rural']
means = [mean_global, mean_urbain, mean_rural]

# Affichage des valeurs dans la console
print(f"Revenu moyen global : {mean_global:,.2f} DH")
print(f"Revenu moyen urbain : {mean_urbain:,.2f} DH")
print(f"Revenu moyen rural : {mean_rural:,.2f} DH")

# Bar‐plot des moyennes
plt.figure(figsize=(6,4))
sns.barplot(x=areas, y=means)
plt.title("Revenu annuel moyen par zone")
plt.ylabel("Revenu moyen (DH/an)")
plt.ylim(0, max(means)*1.1)  # un peu d’espace au‑dessus
for i, m in enumerate(means):
    plt.text(i, m + max(means)*0.02, f"{m:,.0f}", ha='center')
plt.show()
```

```python
# 1) Calcul des moyennes
mean_global = df['income'].mean()
mean_urbain = df[df['area']=='urbain']['income'].mean()
mean_rural  = df[df['area']=='rural']['income'].mean()

# 2) Calcul des pourcentages < moyenne
pct_global = (df['income'] < mean_global).mean() * 100
pct_urbain = (df[df['area']=='urbain']['income'] < mean_urbain).mean() * 100
pct_rural  = (df[df['area']=='rural']['income'] < mean_rural).mean() * 100

# 3) Affichage
print(f"Répartition des revenus < moyenne :")
print(f" • Global : {pct_global:.1f}% (cible 71,8%)")
print(f" • Urbain : {pct_urbain:.1f}% (cible 65,9%)")
print(f" • Rural  : {pct_rural:.1f}% (cible 85,4%)")

# 4) Visualisation
areas = ['Global', 'Urbain', 'Rural']
values = [pct_global, pct_urbain, pct_rural]

plt.figure(figsize=(6,4))
sns.barplot(x=areas, y=values)
plt.ylim(0,100)
plt.ylabel(" % de ménages < moyenne")
plt.title("Répartition des revenus vs % sous la moyenne")
for i,v in enumerate(values):
    plt.text(i, v+2, f"{v:.1f}%", ha='center')
plt.show()
```

```python
# In a new cell

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
df = pd.read_csv("dataset_revenu_marocains.csv")

# Définir l'ordre souhaité des groupes socio-professionnels
group_order = ["Groupe1", "Groupe2", "Groupe3", "Groupe4", "Groupe5", "Groupe6"]

# Calculer le revenu moyen par groupe socio-professionnel
revenu_par_groupe = df.groupby("socio_professional_group")["income"].mean().reindex(group_order)

# Créer le graphique
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x=revenu_par_groupe.index, y=revenu_par_groupe.values, order=group_order, palette="viridis")

# Ajouter les titres et les étiquettes
plt.title("Revenu Annuel Moyen par Groupe Socio-Professionnel", fontsize=16)
plt.xlabel("Groupe Socio-Professionnel", fontsize=12)
plt.ylabel("Revenu Annuel Moyen (DH)", fontsize=12)
plt.xticks(rotation=45, ha="right") # Rotation pour une meilleure lisibilité si les noms sont longs
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajouter les valeurs exactes au-dessus des barres
for index, value in enumerate(revenu_par_groupe.values):
    if pd.notnull(value): # S'assurer que la valeur n'est pas NaN avant de formater
        plt.text(index, value + (revenu_par_groupe.values.max() * 0.01), f"{value:,.0f} DH", ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# In a new cell

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
df_analysis = pd.read_csv("dataset_revenu_marocains.csv")

# Calculer le revenu moyen par genre et état matrimonial
avg_income_marital_gender = df_analysis.groupby(['gender', 'marital_status'])['income'].mean().unstack()

print("Revenu moyen par genre et état matrimonial:")
print(avg_income_marital_gender)

# Plotting général
avg_income_marital_gender.plot(kind='bar', figsize=(12, 7), colormap='viridis')
plt.title('Revenu Moyen par Genre et État Matrimonial', fontsize=16)
plt.ylabel('Revenu Moyen Annuel (DH)', fontsize=12)
plt.xlabel('Genre', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='État Matrimonial', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout(rect=[0, 0, 0.85, 1]) # Adjust layout to make space for legend
plt.show()

# Plotting spécifique pour les hommes
men_marital_order = ['célibataire', 'marié', 'divorcé', 'veuf']
if 'homme' in avg_income_marital_gender.index:
    men_marital_income = avg_income_marital_gender.loc['homme'].reindex(men_marital_order)
    print("\nRevenu moyen des hommes par état matrimonial:")
    print(men_marital_income)
    
    plt.figure(figsize=(8, 5))
    sns.barplot(x=men_marital_income.index, y=men_marital_income.values, order=men_marital_order, palette="Blues_d")
    plt.title('Revenu Moyen des Hommes par État Matrimonial', fontsize=14)
    plt.ylabel('Revenu Moyen Annuel (DH)', fontsize=12)
    plt.xlabel('État Matrimonial', fontsize=12)
    for i, v in enumerate(men_marital_income.values):
        if pd.notnull(v):
            plt.text(i, v + men_marital_income.max()*0.01, f"{v:,.0f}", ha='center', va='bottom')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Plotting spécifique pour les femmes
women_marital_order = ['célibataire', 'marié', 'divorcé', 'veuf']
if 'femme' in avg_income_marital_gender.index:
    women_marital_income = avg_income_marital_gender.loc['femme'].reindex(women_marital_order)
    print("\nRevenu moyen des femmes par état matrimonial:")
    print(women_marital_income)

    plt.figure(figsize=(8, 5))
    sns.barplot(x=women_marital_income.index, y=women_marital_income.values, order=women_marital_order, palette="Oranges_d")
    plt.title('Revenu Moyen des Femmes par État Matrimonial', fontsize=14)
    plt.ylabel('Revenu Moyen Annuel (DH)', fontsize=12)
    plt.xlabel('État Matrimonial', fontsize=12)
    for i, v in enumerate(women_marital_income.values):
        if pd.notnull(v):
            plt.text(i, v + women_marital_income.max()*0.01, f"{v:,.0f}", ha='center', va='bottom')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```
