```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
df = pd.read_csv("dataset_revenu_marocains.csv")

# Calcul des moyennes
mean_global = df['income'].mean()
mean_urbain = df[df['area'] == 'urbain']['income'].mean()
mean_rural = df[df['area'] == 'rural']['income'].mean()

# Préparer les résultats pour le graphique
areas = ['Global', 'Urbain', 'Rural']
means = [mean_global, mean_urbain, mean_rural]

# Affichage des valeurs dans la console
print(f"Revenu moyen global : {mean_global:,.2f} DH")
print(f"Revenu moyen urbain : {mean_urbain:,.2f} DH")
print(f"Revenu moyen rural : {mean_rural:,.2f} DH")

# Bar‐plot des moyennes
plt.figure(figsize=(6,4))
sns.barplot(x=areas, y=means)
plt.title("Revenu annuel moyen par zone")
plt.ylabel("Revenu moyen (DH/an)")
plt.ylim(0, max(means)*1.1)  # un peu d’espace au‑dessus
for i, m in enumerate(means):
    plt.text(i, m + max(means)*0.02, f"{m:,.0f}", ha='center')
plt.show()
```

```python
# 1) Calcul des moyennes
mean_global = df['income'].mean()
mean_urbain = df[df['area']=='urbain']['income'].mean()
mean_rural  = df[df['area']=='rural']['income'].mean()

# 2) Calcul des pourcentages < moyenne
pct_global = (df['income'] < mean_global).mean() * 100
pct_urbain = (df[df['area']=='urbain']['income'] < mean_urbain).mean() * 100
pct_rural  = (df[df['area']=='rural']['income'] < mean_rural).mean() * 100

# 3) Affichage
print(f"Répartition des revenus < moyenne :")
print(f" • Global : {pct_global:.1f}% (cible 71,8%)")
print(f" • Urbain : {pct_urbain:.1f}% (cible 65,9%)")
print(f" • Rural  : {pct_rural:.1f}% (cible 85,4%)")

# 4) Visualisation
areas = ['Global', 'Urbain', 'Rural']
values = [pct_global, pct_urbain, pct_rural]

plt.figure(figsize=(6,4))
sns.barplot(x=areas, y=values)
plt.ylim(0,100)
plt.ylabel(" % de ménages < moyenne")
plt.title("Répartition des revenus vs % sous la moyenne")
for i,v in enumerate(values):
    plt.text(i, v+2, f"{v:.1f}%", ha='center')
plt.show()
```

```python


# Calcul du revenu moyen par âge
age_means = df.groupby('age')['income'].mean()

# Tracé
plt.figure(figsize=(10,6))
plt.plot(age_means.index, age_means.values)
plt.title("Revenu moyen annuel par âge")
plt.xlabel("Âge (années)")
plt.ylabel("Revenu moyen annuel (DH)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("dataset_revenu_marocains.csv")

# Regrouper par âge et calculer le revenu moyen
revenu_par_age = df.groupby("age")["income"].mean()

plt.figure(figsize=(10, 5))
plt.plot(revenu_par_age.index, revenu_par_age.values, color='blue')
plt.title("Revenu moyen en fonction de l'âge")
plt.xlabel("Âge")
plt.ylabel("Revenu annuel moyen (MAD)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Ensure 'df' is defined
if 'df' not in locals():
	df = pd.read_csv("dataset_revenu_marocains.csv")

revenu_par_categorie = df.groupby("age_category")["income"].mean()
categories_ordre = ["jeune", "adulte", "sénior", "agé"]

# Ensure all categories in categories_ordre exist in revenu_par_categorie
categories_ordre = [cat for cat in categories_ordre if cat in revenu_par_categorie.index]

plt.figure(figsize=(7, 5))
plt.bar(categories_ordre, [revenu_par_categorie[cat] for cat in categories_ordre], color='green')
plt.title("Revenu moyen selon la catégorie d'âge")
plt.xlabel("Catégorie d'âge")
plt.ylabel("Revenu annuel moyen (MAD)")
plt.tight_layout()
plt.show()
```

```python
# Correcting the column name from 'milieu' to 'area' and 'revenu_annuel' to 'income'
revenu_par_milieu = df.groupby("area")["income"].mean()

plt.figure(figsize=(6, 4))
plt.bar(revenu_par_milieu.index, revenu_par_milieu.values, color=['orange', 'brown'])
plt.title("Revenu moyen selon le milieu")
plt.xlabel("Milieu")
plt.ylabel("Revenu annuel moyen (MAD)")
plt.tight_layout()
plt.show()
```

```python
# Correcting the column name from 'sexe' to 'gender' and 'revenu_annuel' to 'income'
revenu_par_sexe = df.groupby("gender")["income"].mean()

plt.figure(figsize=(6, 4))
plt.bar(revenu_par_sexe.index, revenu_par_sexe.values, color=['pink', 'skyblue'])
plt.title("Revenu moyen selon le sexe")
plt.xlabel("Sexe")
plt.ylabel("Revenu annuel moyen (MAD)")
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("dataset_revenu_marocains.csv")

ordre_niveaux = ["sans niveau", "fondamental", "secondaire", "supérieur"]
revenu_par_niveau = df.groupby("education")["income"].mean()

plt.figure(figsize=(8, 5))
plt.bar(ordre_niveaux, [revenu_par_niveau.get(niv, 0) for niv in ordre_niveaux], color='purple')
plt.title("Revenu moyen par niveau d'éducation")
plt.xlabel("Niveau d'éducation")
plt.ylabel("Revenu annuel moyen (MAD)")
plt.tight_layout()
plt.show()
```

```python
revenu_par_experience = df.groupby("years_experience")["income"].mean()

plt.figure(figsize=(10, 5))
plt.plot(revenu_par_experience.index, revenu_par_experience.values, color='darkcyan')
plt.title("Revenu moyen en fonction des années d'expérience")
plt.xlabel("Années d'expérience")
plt.ylabel("Revenu annuel moyen (MAD)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
