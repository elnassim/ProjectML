```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
df = pd.read_csv("dataset_revenu_marocains.csv")

# Calcul des moyennes
mean_global = df['Revenu_Annuel'].mean()
mean_urbain = df[df['Milieu'] == 'Urbain']['Revenu_Annuel'].mean()
mean_rural = df[df['Milieu'] == 'Rural']['Revenu_Annuel'].mean()

# Préparer les résultats pour le graphique
areas = ['Global', 'Urbain', 'Rural']
means = [mean_global, mean_urbain, mean_rural]

# Affichage des valeurs dans la console
print(f"Revenu moyen global : {mean_global:,.2f} DH")
print(f"Revenu moyen urbain : {mean_urbain:,.2f} DH")
print(f"Revenu moyen rural : {mean_rural:,.2f} DH")

# Bar‐plot des moyennes
plt.figure(figsize=(6,4))
sns.barplot(x=areas, y=means)
plt.title("Revenu annuel moyen par zone")
plt.ylabel("Revenu moyen (DH/an)")
plt.ylim(0, max(means)*1.1)  # un peu d’espace au‑dessus
for i, m in enumerate(means):
    plt.text(i, m + max(means)*0.02, f"{m:,.0f}", ha='center')
plt.show()
```

```python
# 1) Calcul des moyennes
mean_global = df['Revenu_Annuel'].mean()
mean_urbain = df[df['Milieu']=='Urbain']['Revenu_Annuel'].mean()
mean_rural  = df[df['Milieu']=='Rural']['Revenu_Annuel'].mean()

# 2) Calcul des pourcentages < moyenne
pct_global = (df['Revenu_Annuel'] < mean_global).mean() * 100
pct_urbain = (df[df['Milieu']=='Urbain']['Revenu_Annuel'] < mean_urbain).mean() * 100
pct_rural  = (df[df['Milieu']=='Rural']['Revenu_Annuel'] < mean_rural).mean() * 100

# 3) Affichage
print(f"Répartition des revenus < moyenne :")
print(f" • Global : {pct_global:.1f}% (cible 71,8%)")
print(f" • Urbain : {pct_urbain:.1f}% (cible 65,9%)")
print(f" • Rural  : {pct_rural:.1f}% (cible 85,4%)")

# 4) Visualisation
areas = ['Global', 'Urbain', 'Rural']
values = [pct_global, pct_urbain, pct_rural]

plt.figure(figsize=(6,4))
sns.barplot(x=areas, y=values)
plt.ylim(0,100)
plt.ylabel(" % de ménages < moyenne")
plt.title("Répartition des revenus vs % sous la moyenne")
for i,v in enumerate(values):
    plt.text(i, v+2, f"{v:.1f}%", ha='center')
plt.show()
```

```python
# In a new cell

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
df = pd.read_csv("dataset_revenu_marocains.csv")

# Définir l'ordre souhaité des groupes socio-professionnels
group_order = ["Cadres supérieurs", "Professions intermédiaires", "Employés", "Ouvriers", "Agriculteurs", "Inactifs"]

# Calculer le revenu moyen par groupe socio-professionnel
revenu_par_groupe = df.groupby("CSP")["Revenu_Annuel"].mean().reindex(group_order)

# Créer le graphique
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x=revenu_par_groupe.index, y=revenu_par_groupe.values, order=group_order, palette="viridis")

# Ajouter les titres et les étiquettes
plt.title("Revenu Annuel Moyen par Groupe Socio-Professionnel", fontsize=16)
plt.xlabel("Groupe Socio-Professionnel", fontsize=12)
plt.ylabel("Revenu Annuel Moyen (DH)", fontsize=12)
plt.xticks(rotation=45, ha="right") # Rotation pour une meilleure lisibilité si les noms sont longs
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajouter les valeurs exactes au-dessus des barres
for index, value in enumerate(revenu_par_groupe.values):
    if pd.notnull(value): # S'assurer que la valeur n'est pas NaN avant de formater
        plt.text(index, value + (revenu_par_groupe.values.max() * 0.01), f"{value:,.0f} DH", ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
df = pd.read_csv("dataset_revenu_marocains.csv")

# Convertir home_ownership en binaire

df['total_possessions'] = df['Vehicule_motorise'] + df['Propriete_immobiliere'] + df['Terrain_agricole']

# 1. Graphique des revenus moyens par nombre de possessions
plt.figure(figsize=(12, 6))
sns.barplot(x='total_possessions', y='Revenu_Annuel', data=df, palette='viridis')
plt.title('Revenu Moyen par Nombre de Possessions', fontsize=14)
plt.xlabel('Nombre de Possessions', fontsize=12)
plt.ylabel('Revenu Moyen Annuel (DH)', fontsize=12)
plt.xticks([0, 1, 2, 3], ['0', '1', '2', '3'])
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajouter les valeurs sur les barres
for i, v in enumerate(df.groupby('total_possessions')['Revenu_Annuel'].mean()):
    plt.text(i, v + df['Revenu_Annuel'].mean()*0.01, f"{v:,.0f}", ha='center', va='bottom')

plt.tight_layout()
plt.show()

# 2. Graphique des revenus moyens par type de possession
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))

# Voiture
sns.barplot(x='Vehicule_motorise', y='Revenu_Annuel', data=df, ax=ax1, palette='Blues_d')
ax1.set_title('Revenu Moyen par Possession de Voiture', fontsize=12)
ax1.set_xlabel('Possession de Voiture', fontsize=10)
ax1.set_ylabel('Revenu Moyen Annuel (DH)', fontsize=10)
ax1.set_xticklabels(['Non', 'Oui'])
for i, v in enumerate(df.groupby('Vehicule_motorise')['Revenu_Annuel'].mean()):
    ax1.text(i, v + df['Revenu_Annuel'].mean()*0.01, f"{v:,.0f}", ha='center', va='bottom')

# Propriété immobilière
sns.barplot(x='Propriete_immobiliere', y='Revenu_Annuel', data=df, ax=ax2, palette='Greens_d')
ax2.set_title('Revenu Moyen par Type de Logement', fontsize=12)
ax2.set_xlabel('Type de Logement', fontsize=10)
ax2.set_ylabel('Revenu Moyen Annuel (DH)', fontsize=10)
for i, v in enumerate(df.groupby('Propriete_immobiliere')['Revenu_Annuel'].mean()):
    ax2.text(i, v + df['Revenu_Annuel'].mean()*0.01, f"{v:,.0f}", ha='center', va='bottom')

# Terrain
sns.barplot(x='Terrain_agricole', y='Revenu_Annuel', data=df, ax=ax3, palette='Oranges_d')
ax3.set_title('Revenu Moyen par Possession de Terrain', fontsize=12)
ax3.set_xlabel('Possession de Terrain', fontsize=10)
ax3.set_ylabel('Revenu Moyen Annuel (DH)', fontsize=10)
ax3.set_xticklabels(['Non', 'Oui'])
for i, v in enumerate(df.groupby('Terrain_agricole')['Revenu_Annuel'].mean()):
    ax3.text(i, v + df['Revenu_Annuel'].mean()*0.01, f"{v:,.0f}", ha='center', va='bottom')

plt.tight_layout()
plt.show()

# 3. Graphique des combinaisons de possessions
plt.figure(figsize=(12, 6))
possession_combinations = df.groupby(['Vehicule_motorise', 'Propriete_immobiliere', 'Terrain_agricole'])['Revenu_Annuel'].mean().reset_index()
possession_combinations['combination'] = possession_combinations.apply(
    lambda x: f"Voiture: {'Oui' if x['Vehicule_motorise'] else 'Non'}, "
              f"Logement: {'Oui' if x['Propriete_immobiliere'] else 'Non'}, "
              f"Terrain: {'Oui' if x['Terrain_agricole'] else 'Non'}", axis=1
)

sns.barplot(x='combination', y='Revenu_Annuel', data=possession_combinations, palette='viridis')
plt.title('Revenu Moyen par Combinaison de Possessions', fontsize=14)
plt.xlabel('Combinaison de Possessions', fontsize=12)
plt.ylabel('Revenu Moyen Annuel (DH)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajouter les valeurs sur les barres
for i, v in enumerate(possession_combinations['Revenu_Annuel']):
    plt.text(i, v + df['Revenu_Annuel'].mean()*0.01, f"{v:,.0f}", ha='center', va='bottom')

plt.tight_layout()
plt.show()

# 4. Statistiques détaillées
print("\nStatistiques détaillées par nombre de possessions:")
stats_by_possessions = df.groupby('total_possessions')['Revenu_Annuel'].agg(['mean', 'std', 'min', 'max', 'count'])
print(stats_by_possessions)

print("\nStatistiques détaillées par type de possession:")
print("\nVoiture:")
print(df.groupby('Vehicule_motorise')['Revenu_Annuel'].agg(['mean', 'std', 'min', 'max', 'count']))
print("\nPropriété immobilière:")
print(df.groupby('Propriete_immobiliere')['Revenu_Annuel'].agg(['mean', 'std', 'min', 'max', 'count']))
print("\nTerrain:")
print(df.groupby('Terrain_agricole')['Revenu_Annuel'].agg(['mean', 'std', 'min', 'max', 'count']))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
df = pd.read_csv("dataset_revenu_marocains.csv")

# Calculer le nombre total de possessions
df['total_possessions'] = df['Vehicule_motorise'] + df['Propriete_immobiliere'] + df['Terrain_agricole']

# 1. Graphique des revenus moyens par secteur d'emploi
plt.figure(figsize=(10, 6))
sns.barplot(x='Secteur_emploi', y='Revenu_Annuel', data=df, 
            order=['Privé ', 'Public', 'Informel'],
            palette='viridis')
plt.title('Revenu Moyen par Secteur d\'Emploi', fontsize=14)
plt.xlabel('Secteur d\'Emploi', fontsize=12)
plt.ylabel('Revenu Moyen Annuel (DH)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajouter les valeurs sur les barres
for i, v in enumerate(df.groupby('Secteur_emploi')['Revenu_Annuel'].mean()):
    plt.text(i, v + df['Revenu_Annuel'].mean()*0.01, f"{v:,.0f}", ha='center', va='bottom')

plt.tight_layout()
plt.show()

# 2. Graphique détaillé avec boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Secteur_emploi', y='Revenu_Annuel', data=df,
            order=['Privé ', 'Public', 'Informel'],
            palette='viridis')
plt.title('Distribution des Revenus par Secteur d\'Emploi', fontsize=14)
plt.xlabel('Secteur d\'Emploi', fontsize=12)
plt.ylabel('Revenu Annuel (DH)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajouter les statistiques
stats = df.groupby('Secteur_emploi')['Revenu_Annuel'].agg(['mean', 'median', 'std', 'count'])
print("\nStatistiques détaillées par secteur d'emploi:")
print(stats)

# 3. Graphique de comparaison secteur d'emploi vs possessions
plt.figure(figsize=(15, 6))
sns.boxplot(x='Secteur_emploi', y='Revenu_Annuel', hue='total_possessions', data=df,
            order=['Privé ', 'Public', 'Informel'],
            palette='viridis')
plt.title('Distribution des Revenus par Secteur d\'Emploi et Nombre de Possessions', fontsize=14)
plt.xlabel('Secteur d\'Emploi', fontsize=12)
plt.ylabel('Revenu Annuel (DH)', fontsize=12)
plt.legend(title='Nombre de Possessions')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 4. Statistiques détaillées par secteur
print("\nStatistiques détaillées par secteur d'emploi:")
sector_stats = df.groupby('Secteur_emploi')['Revenu_Annuel'].agg(['mean', 'std', 'min', 'max', 'count'])
print(sector_stats)

# Calculer les différences de revenu moyen entre les secteurs
mean_income_by_sector = df.groupby('Secteur_emploi')['Revenu_Annuel'].mean()
print("\nDifférences de revenu moyen entre les secteurs:")
print(f"Privé vs Public: {((mean_income_by_sector['Privé '] / mean_income_by_sector['Public'] - 1) * 100):.1f}% plus élevé")
print(f"Public vs Informel: {((mean_income_by_sector['Public'] / mean_income_by_sector['Informel'] - 1) * 100):.1f}% plus élevé")
print(f"Privé vs Informel: {((mean_income_by_sector['Privé '] / mean_income_by_sector['Informel'] - 1) * 100):.1f}% plus élevé")

# 5. Graphique des revenus moyens par secteur et zone (urbain/rural)
plt.figure(figsize=(12, 6))
sns.barplot(x='Secteur_emploi', y='Revenu_Annuel', hue='Milieu', data=df,
            order=['Privé ', 'Public', 'Informel'],
            palette='viridis')
plt.title('Revenu Moyen par Secteur d\'Emploi et Zone', fontsize=14)
plt.xlabel('Secteur d\'Emploi', fontsize=12)
plt.ylabel('Revenu Moyen Annuel (DH)', fontsize=12)
plt.legend(title='Zone')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 6. Statistiques par secteur et zone
print("\nStatistiques par secteur d'emploi et zone:")
sector_area_stats = df.groupby(['Secteur_emploi', 'Milieu'])['Revenu_Annuel'].agg(['mean', 'count'])
print(sector_area_stats)
```
